```julia
using PGFPlots
```

```julia
A = [1 1 0 1;1 1 1 0; 0 1 1 1; 1 0 1 1]/3;
T = 10;

x = randn(4,T+1);
ave = sum(x[:,1])/4

for k = 2:T+1
    x[:,k]= A*x[:,k-1];
end


a = Axis([
    Plots.Linear(0:T, x[1,:], legendentry=L"$x_1$",mark="none"),
    Plots.Linear(0:T, x[2,:], legendentry=L"$x_2$",mark="none"),
    Plots.Linear(0:T, x[3,:], legendentry=L"$x_3$",mark="none"),
    Plots.Linear(0:T, x[4,:], legendentry=L"$x_4$",mark="none"),
    Plots.Linear([0,T], [ave,ave], style="dashed",mark="none"),
],
    xlabel=L"$k$",ylabel=L"$x_i(k)$",)
```

```julia
v = randn(4,T+1);

for k = 2:T+1
    x[:,k]= A*(x[:,k-1]+v[:,k]);
end


b = Axis([
    Plots.Linear(0:T, x[1,:], legendentry=L"$x_1$",mark="none"),
    Plots.Linear(0:T, x[2,:], legendentry=L"$x_2$",mark="none"),
    Plots.Linear(0:T, x[3,:], legendentry=L"$x_3$",mark="none"),
    Plots.Linear(0:T, x[4,:], legendentry=L"$x_4$",mark="none"),
    Plots.Linear([0,T], [ave,ave], style="dashed",mark="none"),
],
    xlabel=L"$k$",ylabel=L"$x_i(k)$",)
```

```julia
v = randn(4,T+1);
φ = 0.7;
for k = 2:T+1
    x[:,k]= A*(x[:,k-1]+φ^(k-2).*v[:,k-1]);
end

c = Axis([
    Plots.Linear(0:T, x[1,:], legendentry=L"$x_1$",mark="none"),
    Plots.Linear(0:T, x[2,:], legendentry=L"$x_2$",mark="none"),
    Plots.Linear(0:T, x[3,:], legendentry=L"$x_3$",mark="none"),
    Plots.Linear(0:T, x[4,:], legendentry=L"$x_4$",mark="none"),
    Plots.Linear([0,T], [ave,ave], style="dashed",mark="none"),
],
    xlabel=L"$k$",ylabel=L"$x_i(k)$",)
```

```julia
v = randn(4,T+1);
φ = 0.7;

old = zeros(4,1);
new = zeros(4,1);
for k = 2:T+1
    new = φ^(k-2).*v[:,k];
    x[:,k]= A*(x[:,k-1]+new-old);
    old = new
end

d = Axis([
    Plots.Linear(0:T, x[1,:], legendentry=L"$x_1$",mark="none"),
    Plots.Linear(0:T, x[2,:], legendentry=L"$x_2$",mark="none"),
    Plots.Linear(0:T, x[3,:], legendentry=L"$x_3$",mark="none"),
    Plots.Linear(0:T, x[4,:], legendentry=L"$x_4$",mark="none"),
    Plots.Linear([0,T], [ave,ave], style="dashed",mark="none"),
],
    xlabel=L"$k$",ylabel=L"$x_i(k)$",)
```

```julia
save("consensus1.tex",a)
save("consensus2.tex",b)
save("consensus3.tex",c)
save("consensus4.tex",d)
```
